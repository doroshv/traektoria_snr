## Введение

Взрыв Галактической сверхновой - одно из самых ожидаемых и вероятно впечатляющих событий для любого астронома! Одно из таких событий произошло в 1572 г и часто назывыется взрывом сверхновой Тихо (Tycho). Не потому что Тихо Браге был первым или единственным наблюдателем увидевшим новую звезду (все на земле видели!), но потому что он был первым кто детально описал свои (и чужие) наблюдения в труде “De nova et nullius aevi memoria prius visa stella” (“О звезде новой и никогда ранее никем не виданной в жизни или памяти”), который был опубликован в 1573 г.

Браге и другие наблюдатели заметили что новая звезда появилась на Небе 2-3 ноября, и к 11 ноябрю стала ярче чем Юпитер. Пик яркости был достигнут к 16 ноябрю 1572г когда звездная величина оценивалась $m_v\sim−4.0$ (сравнимо с максимальной яркостью Венеры). После пика сверховая стала постепенно ослабевать но оставалась видимой невооруженным глазом до начала 1574г. Остатками сверхновой (supernova remnant или SNR) называют газовые структуры остающиеся после взрывоав сверхновых. Внешний край такого остата ограничен расширяющейся ударной волной и заполнен горячим расширяющимся веществом выброшенным во время взрыва и межзвездным веществом прошедшим через ударную волну. Хотя и не такие яркие как сама сверхновая, остатки сверхновых тоже являются источником излучения в широком диапазоне длин волн (от радио до ТэВ), и играют центральную роль в обогащении межзвездной среды тяжелыми элементами (синтезированными в течении жизни взорвавшейся звезды и во время взрывы), а также в ускорении космических лучей.

Остатки сверхновых не имеют собственных источников энергии и светят только за счет тепловой и механической энергии выделенной во время взрыва. Их эволюция определяется тем насколько эффективно и какими спосабами эта энергия теряется. Например, тепловая энергия выброшенного вещества в основном уносится свободно-свободным рентгеновским излучением (хотя, конечно, остатки также охлаждаются из за расшириния остатка). Механическая энергия ударной волны в основном тратится на нагрев межзвездной среды (что опять же приводит к тепловому излучению в рентгеновском диапазоне), на ускорениее космических лучей, и др. В целом, постепенная потеря энергии приводит к замедлению расширения осстатка свархновой, и анализ и моделирования этого процесса являются прекрасными инструментами для изучения как энергетики и физических процессов в самой сверхновой, ее остатке, а также свойств межзвездной среды. Однако детальное моделирование этих процессов - довольно сложная задача которая выходит далеко за пределы данной лабораторной работы и не является нашей целью сегодня. Мы можем однако, сделать первый шаг в этом направлении и измерить то что мы можем измерить, а именно - темп расшириения остатка. 

## Наблюдательные данные
Сегодня наш анализ будет основан на данных полученнных обсерваторией *Chandra*. Телескопы этоя обсерватории имеют наилучшее пространственное разрешение из всех когда либо построенных рентгеновских телескопов (а нам нужен рентгеновский телескоп, поскольку на ранних стадиях эволюции остатки сверхновых светят в основном в рентгене), а кроме того, *Chandra* наблюдала этот остаток много раз на протяжении долгого времени. Основной идеей нашего сегодняшнего эксперемента является сравнение изображений остатка полученных *Chandra* в нескольких разнесенных по времени наблюдениях с целью измерения темпа увеличения этого радиуса со временем. Измерение темпа расширения позволит нам определить возраст остатка (поскольку дата и время каждого наблюдения известны).

Данные необходимые для того чтобы это сделать можно скачать из [HEASARC data archive](http://heasarc.gsfc.nasa.gov/cgi-bin/W3Browse/w3browse.pl). Проще всего это можно сделать используя ```Browse``` интерфейс архива. Нужно ввести *Tycho SNR* в строке поиска, выбрать *Chandra* в качестве единственной миссии которая нас интересует, и нажать "Start Search". Как только появятся результаты, можно кликнуть на вкладку *chanmaster* чтобы увидеть полный список наблюдений этого источника Чандрой. Имеет смысл начать со сравнения самого первого и самого последнего наблюдения (что соответсвует максимальному изменению линейного размера остатка).

## Анализ радиальных профилей с помощью Python

напрмимер, полученные профили можно проанализировать с помощью *Python*. Поскольку речь идет о работе с данными, удобно воспользоваться библиотекой *numpy* для работы с данными и *matplotlib* (*pylab*) для отображения результатов

````python
import numpy as np
import pylab as pl

r1,f1,f1_err = np.loadtxt('red.dat',unpack=True)[:3] # загрузка первого профиля
r2,f2,f2_err = np.loadtxt('blue.dat',unpack=True)[:3] # загрузка второго профиля

pl.errorbar(r1,f1/max(f1),f1_err/max(f1)) # рисуем первый профиль (нормированный на макс яркость)
pl.errorbar(r2,f2/max(f2),f2_err/max(f2)) # и второй...
pl.xlabel('Radius, as')
pl.ylabel('Norm. intensity')
````

Можно определить сдвиг между двумя профилями визуально, т.е. рисуя их на одном графике и сдвигая один относительно другого пока не совпадут:

````python
pl.errorbar(r1,f1/max(f1),f1_err/max(f1))
pl.plot(r1+5.5,f1/max(f1),'k:')
pl.errorbar(r2,f2/max(f2),f2_err/max(f2))
pl.xlabel('Radius, as')
pl.ylabel('Norm. intensity')
````

Естественно, это мало отличается от того что бы делали ранее, т.е. радиус все еще подбирается "на глаз". Мы можем уйти от этого и определять радиус исходя насколько сдвинутая версия одного наблюдения совпадает со вторым. Для этого удобно интерполировать один из профилей, а затем вычисллять значение $\chi^2$ статистики для любого произвольного сдвига.

````python
# определим модель как интерполированную версию радиального профиля остатка из "старого наблюдения"
model = lambda shift: np.interp(r2,r1+shift,f1/max(f1))
# теперь нормируем данные
data = f2/max(f2)
error = 2*f2_err/max(f2) # тут нужно заметить что наша модель тоже содержит ошибку которую проще всего приписать данным
# пробные сдвиги которые мы ходим проверить
offset = np.linspace(0,10,300)
# считаем отклонение модели от данных
chi2 = [sum(((model(x)-data)/error)**2) for x in offset]
pl.plot(offset,chi2)
# теперь можно определить минимум
print("Optimal shift is %.2f"%(offset[np.argmin(chi2)]))
pl.axvline(offset[np.argmin(chi2)])
````

Конечно, минимизация функции r$\chi^2$ возможна и в автоматическом режиме (i.e. *minimize $\chi^2$*). Для этого момжно воспользоваться модулем *scipy.optimize*.

````python
import scipy.optimize

chi2 = lambda x: sum(((model(x)-data)/error)**2)
best_x = scipy.optimize.fmin(chi2, [5])
print("Optimal offset is %.2f''"%best_x[0])
````

Мы можем перерисовать профили с этим сдвигом:

````python
pl.errorbar(r1,f1/max(f1),f1_err/max(f1))
pl.plot(r1+best_x[0],f1/max(f1),'k:')
pl.errorbar(r2,f2/max(f2),f2_err/max(f2))
pl.xlabel('Radius, as')
pl.ylabel('Norm. intensity')
````

## Оценка возраста остатка
Первым шагом в нашем случае является измерение радиуса для обоих наблюдений разнесенных по времени, т.е. измерение темпа роста радиуса остатка (поскольку промежуток времени между наблюдениями известен). Например,мы получили изменение радиуса $\Delta R\simeq6$'' между наблюдениями. Если посмотреть на даты наблюдений (HEASARC, либо просто по заголовкам FITS файлов), составляет примерно  $\sim14.6$ лет. Это означает что средний темп расширения за время наблюдений составлял около $6/14.6\sim0.4$ сек/год. Для оценки возраста остатка можно предположить что изначально объект был точкой, и поделить текущий радиус на наблюдаемый темп расширения, т.е. момент взрыва произошел  $t_{0} = 257.25/0.38 = 682$ лет назад. В нашем случае мы знаем реальный момент взрыва (сентябрь 1572 г.) и можем сравнить его с нашей оценкой (в которой взрыв произошел $2000 - 682 = 1318$г. Очевидно, что есть некоторое несоответствие, которое указывает на то что реально с момента взрыва остаток в среднем расширялся быстрее чем сейчас. 
Почему так может быть? Мы упомянули вначале что остатки сверхновых светят за счет выделения начальной кинетической и тепловой энергии полученой во взрыве, так что в целом ничего супер удивительного тут нет. ```https://heasarc.gsfc.nasa.gov/FTP/chandra/data/byobsid/5/10095/primary/acisf10095N003_full_img2.fits.gz```
Этот адрес сконструирован по простым правилам и мы можем сделать это имея только идентификатор наблюдения из нашего запроса. Как только мы знаем где лежат данные, мы можем открыть их удаленно с помощью модуля  *astropy.io.fits*. Если понадобится нарисовать картинки в астрономических координатах можно воспользоваться модулем *aplpy*. Давайте сначала сделаем это для нашего первого наблюдения:


````python
import astropy.io.fits as fits
import aplpy

def get_img(obsid):
    obsid = "%05d"%obsid
    try:
        url = 'https://heasarc.gsfc.nasa.gov/FTP/chandra/data/byobsid/%s/%d/primary/acisf%sN003_full_img2.fits.gz'%(obsid[-1],int(obsid),obsid)
        hdu = fits.open(url)
        return hdu
    except:
        pass
    try:
        url = 'https://heasarc.gsfc.nasa.gov/FTP/chandra/data/byobsid/%s/%d/primary/acisf%sN002_full_img2.fits.gz'%(obsid[-1],int(obsid),obsid)
        hdu = fits.open(url)
        return hdu
    except:
        pass
    try:
        url = 'https://heasarc.gsfc.nasa.gov/FTP/chandra/data/byobsid/%s/%d/primary/acisf%sN004_full_img2.fits.gz'%(obsid[-1],int(obsid),obsid)
        hdu = fits.open(url)
        return hdu
    except:
        pass
    return False

# get the earliest image we found previously

earliest_image = get_img(115)
fig = aplpy.FITSFigure(earliest_image)
fig.show_colorscale()
````

Наконец, для того чтобы строить радиальные профили нам нужно знать центр объекта. Это можно сделать на глаз как мы делали раньше, но интереснее попытаться найти центр автоматически. Например с помощью т.н. *circular Hough* преобразования которое используется для поиска окружностей в изображениях. Это преобразование - часть модуля *scikits-image*. Как и большинство методов анализа, оно лучше работает если мы сначала найдем края объектов (в данном случае остатка). Это можно сделать в два этапа: сначала отделяем объект от фона (просто сравнивая значения пикселей), а уже затем ищем края с помощью *scikits-image* [(см так же этот пример)](https://scikit-image.org/docs/stable/auto_examples/edges/plot_circular_elliptical_hough_transform.html#sphx-glr-auto-examples-edges-plot-circular-elliptical-hough-transform-py) here.

````python
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.feature import canny
from skimage.draw import circle_perimeter
from skimage.util import img_as_ubyte
from skimage import data, color

img = img_as_ubyte(im_e>=50)
pl.imshow(im_e)
pl.imshow(img,alpha=0.3)
edges = canny(img, sigma=3)
pl.imshow(edges,alpha=0.3)
````

Теперь можно поискать центр

````python
# from the image we can tell that the radius is ~60 pixels, so we search around this value
hough_radii = np.arange(50, 80, 1)
hough_res = hough_circle(edges, hough_radii)

# find one (main) circle approximating the image. Try to play with total_num_peaks
accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii,total_num_peaks=1)

# copy image of the SNR for visualization
image = color.gray2rgb(2*im_e/im_e.max())

for center_y, center_x, radius in zip(cy, cx, radii):
    circy, circx = circle_perimeter(center_y, center_x, radius,
                                    shape=image.shape)
    image[circy, circx] = (220, 20, 20)

pl.imshow(image)
print("Circle pixel coordinates are %d,%d, and radius=%d pixels"%(center_x,center_y,radii))
````

Теперь мы можем посчитать расстояние до центра для каждого пиксела. Для этого нам нужно знать пиксельные координаты которые мы для начала и сгенирируем. Помните, что если мы хотим перевести радиусы в градусы нам нужно знать размер пиксела (который хранится в заголовке файла и составляет 0.0010933333333334 градуса). 

````python
# note that Numpy and fortran/fits have different convetion regarding the order of indexing, hence, our choice of indices here
X,Y = np.meshgrid(np.arange(im_e.shape[1]),np.arange(im_e.shape[0]))
D = np.sqrt((X-center_x)**2+(Y-center_y)**2)
# show original
pl.imshow(im_e)
# overlayed with distance matrix
pl.imshow(D,alpha=0.5)
````

Тепреь можно построить радиальный профиль для обоих изображений. Для этого нам нужно посчитать среднюю яркость в каждом радиальном бине. Здесь проще всего воспользоваться модулем *scipy.stats*. Обратите внимание что исходные изображения у нас в "отсчетах" а не потоке, т.е. необходимо нормировать оба финальных профиля.

````python
import scipy.stats
rbins = np.linspace(1,120,250)
hist_e = scipy.stats.binned_statistic(D.flatten(),im_e.flatten(),bins=rbins)
hist_l = scipy.stats.binned_statistic(D.flatten(),im_l.flatten(),bins=rbins)
pl.plot(hist_e[1][:-1],(hist_e[0])/np.nanmax(hist_e[0]),'r-',label='early')
pl.plot(hist_l[1][:-1],hist_l[0]/np.nanmax(hist_l[0]),'b-',label='late')
# indicate also the radius we determined from the hough fit
pl.axvline(radii)
pl.legend(frameon=False)
````

Тот факт что изображение в отсчетах ответственен и за то что мы наблюдаем изменения яркости: в реальном инструменте экспозиция (и количество отсчетов) зависят от того куда мы смотрим и должно корректироваться (ключевое слово - карты эккспозиции или "exposure maps"). Тут мы этого не делаем для простоты, но это может повлиять на результаты. Что бы уменьшить этот эффект, сконцентрируемся на внешнем крае остатка, например, между пикселями 45 и 70. 

````python
rbins = np.linspace(45,70,100)
hist_e = scipy.stats.binned_statistic(D.flatten(),im_e.flatten(),bins=rbins,statistic='median')
hist_l = scipy.stats.binned_statistic(D.flatten(),im_l.flatten(),bins=rbins,statistic='median')
pl.plot(hist_e[1][:-1],(hist_e[0])/np.nanmax(hist_e[0]),'r-',label='early')
pl.plot(hist_l[1][:-1],hist_l[0]/np.nanmax(hist_l[0]),'b-',label='late')
# indicate also the radius we determined from the hough fit
pl.axvline(radii)
pl.legend(frameon=False)
````

Опять же, мы видим сдвиг между двумя наблюдениями. Давайте определим его автоматически с помощью *scipy.optimize* 

````python
import scipy.optimize
import astropy.units as u

px2as = 3.936

def fitness_function(p):
    hist_scaled = scipy.stats.binned_statistic(D.flatten()*p[0],im_e.flatten(),bins=rbins,statistic='median')[0]
    return sum((hist_scaled-hist_l[0])**2)

# find optimal scaling allowing to match both profiles optimally
scale_factor = scipy.optimize.fmin(fitness_function,[1.0])

# get conversion factor from pixel to arcseconds

# print it out
print("The old image needs to be scaled by %.2f to match new one"%scale_factor[0])
print("This corresonds to %f'' expansion of the shell"%((radii*scale_factor-radii)*px2as))
print("Assumed shell radius here is %.2f''"%(radii*px2as))

# rbins = np.linspace(45,70,100)
hist_e = scipy.stats.binned_statistic(D.flatten()*scale_factor[0],im_e.flatten(),bins=rbins)
hist_l = scipy.stats.binned_statistic(D.flatten(),im_l.flatten(),bins=rbins)
pl.plot(hist_e[1][:-1],(hist_e[0])/np.nanmax(hist_e[0]),'r-',label='early')
pl.plot(hist_l[1][:-1],hist_l[0]/np.nanmax(hist_l[0]),'b-',label='late')
# indicate also the radius we determined from the hough fit
pl.axvline(radii)
pl.legend(frameon=False)
````

Как вы видите, разница в радиусах довольно сильно отличается от того что мы получили ранее. Основной причиной является то что мы теперь используем значительную часть остатка а не один тонкий филамент на внешнем его края. Для этой части (показанной вертикальной линией $R\sim R_{SNR}$) видно что мы переоценили темп расширения, т.е. внешний край расширяется медленнее чем весь остаток в целом, что в принципе ожидаемо. Если мы выберем $R\sim54$px как характерный радиус части остатка которую мы измеряем, мы можем посчитать возраст остатка для этой его части:

````python
Rmid = 54
# we can now evaluate age of the SNR
expansion_rate = (Rmid*scale_factor-Rmid)/14.6 # arcseconds per year
expansion_rate_as = expansion_rate*px2as
age = Rmid/expansion_rate
print("so, we estimate expansion rate of %.2f''/yr, which implies age of %.1fyr. The true age is 2000-1572=%.1fyr"%(expansion_rate_as,age,2000-1572))
````

Как мы видим, разница с реальным возрастом уменьшилась! Если учесть так же эффекты проекции, то можно получить еще лучшее согласие оценочного и реального возраста для большей части остатка. Самые внешние его части, однако, действительно расширяются медленнее, т.е. мы видим взаимодействие остатка сверхновой с межзвездной средой прямо сейчас!
